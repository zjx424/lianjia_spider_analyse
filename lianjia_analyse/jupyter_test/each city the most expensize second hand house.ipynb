```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
dict={'dg':'东莞','gz':'广州','sz':'深圳','fs':'佛山','zh':'珠海','zs':'中山','hui':'惠州'}
n=1
for i in dict:
    city=dict[i]
    print('{}豪宅top10'.format(city))
    gd_data=pd.read_csv('./gd{}.csv'.format(n))
    gd_list=gd_data.sort_values(by="unit_Price",ascending=False)['address']
    print(gd_list.drop_duplicates().head(10))
    n+=1
```

# 商品房数量

```python
type_list=[]
for i in range(1,8):
    each_data=(pd.read_csv('./gd{}.csv'.format(i)).type=='商品房').sum()
    type_list.append(each_data)
type_list
```

```python
import matplotlib.pyplot as plt
import matplotlib as mpl
mpl.rcParams['font.family']='simhei'
mpl.rcParams['axes.unicode_minus']=False # 处理负号问题 
mpl.rcParams['font.size'] = 20.0
# Pie chart, where the slices will be ordered and plotted counter-clockwise:
labels=[]
for i in dict:
    labels.append(dict[i])
# sizes = [8.13,26.68,31.92,13.34,2.66,5.01,12.27]
sizes=[]
for i in type_list:
    sizes.append(round(i/sum(type_list),2))
# explode = (0, 0.1, 0, 0)  # only "explode" the 2nd slice (i.e. 'Hogs')
fig1, ax1 = plt.subplots(figsize=(12,12),)
ax1.pie(sizes,labels=labels, autopct='%1.1f%%', startangle=60)
ax1.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.


ax1.axis('equal')
plt.text(1.4,0.4,
'''东莞商品房数量为:9694套
广州商品房数量为:31262套
深圳商品房数量为:37460套
佛山商品房数量为:16162套
珠海商品房数量为:3077套
中山商品房数量为:5983套
惠州商品房数量为:14411套''')
plt.show()

```
